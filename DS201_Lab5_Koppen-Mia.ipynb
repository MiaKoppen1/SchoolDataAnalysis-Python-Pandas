```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
#reading the dataset
state_schools = pd.read_csv('lab5.csv') #reading the data inside of the file "Lab.csv"
print(state_schools.head()) #shows every row and column in the chart
print(" ")
print("dimensions of the chart")
print(state_schools.shape) #dimensions (rows and columns)
print(" ")
print("the columns:")
print(state_schools.columns) #shows me the list of columns and their names
print(" ")
print("data types of the columns:")
state_schools.dtypes
```

```python
state_schools2 = state_schools.groupby(['State_name']).size() #grouping the data by state names
type(state_schools2) #prints out the type of state_schools2 (series)
state_schools.describe() #shows statistics
print(state_schools2) #prints out the chart sorted by state names.

colors = ['black']

state_schools2.plot.bar(figsize=(17,4), color=colors);

plt.show() #showing the bar chart to the screen.
```

```python
housing_data = pd.read_csv('Housing.csv')
print(housing_data.head())
```

```python
print(housing_data.columns)
print(" ")
print(housing_data.shape)
print(" ")
print(housing_data.dtypes)
housing_data.info
housing_data.describe()
```

```python

housing_price_mean = housing_data['price'].mean() #adds the mean of the price column to the housing_price_mean variable.
print(housing_price_mean) #prints out the mean of the price column.
bedrooms_max = housing_data['bedrooms'].max() #adds the max number of bedrooms to the bedrooms_max variable.
print(bedrooms_max) #prints out the max of the price column.
```

```python
housing_data['mainroad'].value_counts() #amount of houses are on the main road.
main_road = housing_data.groupby(['mainroad']).size() #grouping the data by state names

main_road.plot.bar(figsize=(4,4), color = 'pink');
plt.show()

```
